```python
import pandas as pd
import numpy as np
import re
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error


df = pd.read_csv("imdb_movies.csv2")
```

```python
df['Votes'] = df['Votes'].apply(lambda x: int(re.sub(r'[^\d]', '', str(x))))


def parse_gross(gross):
    if 'M' in gross:
        return float(re.sub(r'[^\d.]', '', gross)) * 1e6
    elif '$' in gross:
        return float(re.sub(r'[^\d.]', '', gross))
    else:
        try:
            return float(re.sub(r'[^\d.]', '', gross))
        except ValueError:
            return np.nan

df['Gross_collection'] = df['Gross_collection'].apply(parse_gross)
```

```python
feature_columns = ['Votes']


X = df[feature_columns]
y_votes = df['Votes']
y_gross = df['Gross_collection']


X_train, X_test, y_votes_train, y_votes_test = train_test_split(X, y_votes, test_size=0.2, random_state=42)
_, _, y_gross_train, y_gross_test = train_test_split(X, y_gross, test_size=0.2, random_state=42)


votes_model = LinearRegression()
votes_model.fit(X_train, y_votes_train)


gross_model = RandomForestRegressor(n_estimators=100, random_state=42)
gross_model.fit(X_train, y_gross_train)
```

```python
y_votes_pred = votes_model.predict(X_test)
y_gross_pred = gross_model.predict(X_test)


mse_votes = mean_squared_error(y_votes_test, y_votes_pred)
mse_gross = mean_squared_error(y_gross_test, y_gross_pred)

print(f"Mean Squared Error for 'Votes' prediction: {mse_votes:.2f}")
print(f"Mean Squared Error for 'Gross_collection' prediction: {mse_gross:.2f}")
```

```python
import pandas as pd
import numpy as np
import re
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import StandardScaler


df = pd.read_csv("imdb_movies.csv2")
```

```python
df['Votes'] = df['Votes'].apply(lambda x: int(re.sub(r'[^\d]', '', str(x))))
```

```python
def parse_gross(gross):
    if 'M' in gross:
        return float(re.sub(r'[^\d.]', '', gross)) * 1e6
    elif '$' in gross:
        return float(re.sub(r'[^\d.]', '', gross))
    else:
        try:
            return float(re.sub(r'[^\d.]', '', gross))
        except ValueError:
            return np.nan

df['Gross_collection'] = df['Gross_collection'].apply(parse_gross)


feature_columns = ['Votes']
```

```python
X = df[feature_columns]
y_votes = df['Votes']
```

```python
X_train, X_test, y_votes_train, y_votes_test = train_test_split(X, y_votes, test_size=0.2, random_state=42)


scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
param_grid = {
    'n_estimators': [50, 100, 150],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}
```

```python
grid_search = GridSearchCV(estimator=RandomForestRegressor(random_state=42),
                           param_grid=param_grid,
                           scoring='neg_mean_squared_error',
                           cv=5,
                           verbose=2)

grid_search.fit(X_train_scaled, y_votes_train)


best_params = grid_search.best_params_
print("Best Parameters:", best_params)


best_votes_model = RandomForestRegressor(random_state=42, **best_params)
best_votes_model.fit(X_train_scaled, y_votes_train)


y_votes_pred = best_votes_model.predict(X_test_scaled)


mse_votes = mean_squared_error(y_votes_test, y_votes_pred)
print(f"Mean Squared Error for 'Votes' prediction: {mse_votes:.2f}")


mse_votes = mean_squared_error(y_votes_test, y_votes_pred)
print(f"Mean Squared Error for 'Gross_collection' prediction: {mse_votes:.2f}")
```

```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import StandardScaler
import joblib
```

```python
df = pd.read_csv("imdb_movies.csv2")
```

```python
df['Votes'] = df['Votes'].apply(lambda x: int(re.sub(r'[^\d]', '', str(x))))
```

```python
def parse_gross(gross):
    if 'M' in gross:
        return float(re.sub(r'[^\d.]', '', gross)) * 1e6
    elif '$' in gross:
        return float(re.sub(r'[^\d.]', '', gross))
    else:
        try:
            return float(re.sub(r'[^\d.]', '', gross))
        except ValueError:
            return np.nan

df['Gross_collection'] = df['Gross_collection'].apply(parse_gross)


feature_columns = ['Votes']
```

```python
X = df[feature_columns]
y_votes = df['Votes']
y_gross = df['Gross_collection']
X_train, X_test, y_votes_train, y_votes_test, y_gross_train, y_gross_test = train_test_split(X, y_votes, y_gross, test_size=0.2, random_state=42)


scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)


param_grid = {
    'n_estimators': [50, 100, 150],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

grid_search = GridSearchCV(estimator=RandomForestRegressor(random_state=42),
                           param_grid=param_grid,
                           scoring='neg_mean_squared_error',
                           cv=5,
                           verbose=2)

grid_search.fit(X_train_scaled, y_votes_train)
```

```python
best_params = grid_search.best_params_
print("Best Parameters:", best_params)


best_votes_model = RandomForestRegressor(random_state=42, **best_params)
best_votes_model.fit(X_train_scaled, y_votes_train)


y_votes_pred = best_votes_model.predict(X_test_scaled)


mse_votes = mean_squared_error(y_votes_test, y_votes_pred)
print(f"Mean Squared Error for 'Votes' prediction: {mse_votes:.2f}")


feature_importances_votes = best_votes_model.feature_importances_


gross_model = RandomForestRegressor(random_state=42)
gross_model.fit(X_train_scaled, y_gross_train)


y_gross_pred = gross_model.predict(X_test_scaled)


mse_gross = mean_squared_error(y_gross_test, y_gross_pred)
print(f"Mean Squared Error for 'Gross_collection' prediction: {mse_gross:.2f}")


feature_importances_gross = gross_model.feature_importances_
print("Feature Importances for 'Votes' model:", feature_importances_votes)
print("Feature Importances for 'Gross_collection' model:", feature_importances_gross)
```
